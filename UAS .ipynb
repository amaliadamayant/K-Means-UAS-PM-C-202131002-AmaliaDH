```python
df.columns = df.columns.str.strip()
```

```python
x = df[['Berat','Tinggi']].values
```

```python
from sklearn.preprocessing import MinMaxScaler 
scaler = MinMaxScaler()
x = scaler.fit_transform(x)
```

```python
x
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=4, random_state=0)
kmeans.fit(x)
```

```python
y_cluster = kmeans.fit_predict(x)
```

```python
y_cluster
```

```python
df['Cluster'] = y_cluster
```

```python
df.head(10)
```

```python
kmeans.cluster_centers_
```

```python
plt.figure(figsize=(20, 10))
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_)
plt.scatter(kmeans.cluster_centers_[0,0],
            kmeans.cluster_centers_[0,1],
            cmap='rainbow',marker='*',
            s=150,label='Centroid 0')
plt.scatter(kmeans.cluster_centers_[1,0],
            kmeans.cluster_centers_[1,1],
            cmap='rainbow',marker='>',
            s=150,label='Centroid 1')
plt.scatter(kmeans.cluster_centers_[2,0],
            kmeans.cluster_centers_[2,1],
            cmap='rainbow',marker='<',
            s=150,label='Centroid 1')
plt.scatter(kmeans.cluster_centers_[3,0],
            kmeans.cluster_centers_[3,1],
            cmap='rainbow',marker='^',
            s=150,label='Centroid 1')
plt.xlabel("Berat")
plt.ylabel("Tinggi")
plt.legend()
```

```python
from sklearn.metrics import silhouette_score

silhouette_avg = silhouette_score(x, kmeans.labels_)
print("Silhouette Coefficient:", silhouette_avg)
```

```python
from sklearn.cluster import KMeans
cs = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i)
    kmeans.fit(x)
    cs.append(kmeans.inertia_)
cs
```

```python
plt.plot(range(1, 11), cs)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('CS')
plt.show()
```
